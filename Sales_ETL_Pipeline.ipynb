### Feature Engineering — Customer Name & Paid Status

```python
df["Customer Name"] = df["Customer First Name"] + " " + df["Customer Last Name"]
```

```python
df.drop(columns=["Customer First Name","Customer Last Name"], inplace=True)
df
```

```python
def Paid_Status(row):
    if row["Paid"]>0:
        return "Paid"
    else:
        return "Not Paid"
df["Paid_Status"] = df.apply(Paid_Status, axis=1)
df
```

### Rename Columns — Replace Spaces with Underscores

```python
df.columns = df.columns.str.replace(' ', '_')
df.columns
```

## Orders Table

```python
df = pd.read_csv("E:\Orders.csv")
```

### Explore

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

### Fix Shipping Cost — Convert to Numeric

```python
df['Shipping Cost'] = pd.to_numeric(df['Shipping Cost'], errors='coerce')
print(df['Shipping Cost'].max(), df['Shipping Cost'].min())
```

### Fill Missing Values

```python
df['Shipping Cost'].fillna(df['Shipping Cost'].mean(), inplace=True)
df['Status'].fillna('Pending', inplace=True)
```

```python
df.isnull().sum()
```

### Rename Columns — Replace Spaces with Underscores

```python
df.columns = df.columns.str.replace(' ', '_')
df.columns
```

### Feature Engineering — Avg Rating by Category

```python
df['Rating'] = df.groupby(df['Product_ID'].apply(lambda x: x.split('-')[0]))['Rating'].transform(lambda x: x.fillna(x.mean()))
```

### Feature Engineering — Net Profit

```python
df['Net_Profit_B'] = df['Profit'] - df['Shipping_Cost']
df['Net_Profit_A'] = df['Net_Profit_B'] - (df['Net_Profit_B'] * df['Discount'])
df.head(20)
```

### Load to SQL Server — Orders

```python
from sqlalchemy import create_engine
import pandas as pd

server   = r'DESKTOP-J0HPDNG\SQLEXPRESS'
database = 'Sales_Insights'
driver   = 'ODBC Driver 18 for SQL Server'

engine = create_engine(
    f"mssql+pyodbc://@{server}/{database}?driver={driver.replace(' ', '+')}&trusted_connection=yes&TrustServerCertificate=yes"
)
```

```python
table_name = "Orders_Data"
df.to_sql(table_name, engine, if_exists="replace", index=False)
pd.read_sql("SELECT TOP 5 * FROM Orders_Data", engine)
```
